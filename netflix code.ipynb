```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('netflix_data.csv')
df.sample(5)
```

```python
df.isnull().sum()
# Hence no null values are present
```

```python
#view the dataset info
print('information of datastet:\n',df.info())

#checkinjg any duplicated rows
print('\nchecking duplicated values present or not :\n',df.duplicated().sum()) # no duplicate rows is present 
```

```python
# creating a new column 'genre' which would include the 'listed_in' item with type of shows and movies it is
df['genre']=df['listed_in'].str.split(',',expand=True)[0]
df['genre'].value_counts()
```

# Data visualizing

```python
print('\n',df['type'].value_counts())

#plotting the show type and movies
plt.subplots(figsize=(15,5))
plt.subplot(121)
df['type'].value_counts().plot(kind='pie',autopct='%.2f%%')
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.title('Type')
plt.subplot(122)
df['type'].value_counts().plot(kind='bar')
plt.title('Type')
plt.show()
```

* TOP 10 RATING ON THE NETFLIX

```python
print('Top Rating on Netflix:\n',df['rating'].value_counts()[0:10])

# plotting the rating chart
plt.subplots(figsize=(15,5))
plt.subplot(121)
df['rating'].value_counts()[:10].plot(kind='bar') # plotting the top 10 rating in bar chart
plt.ylabel('rating')
plt.title('Rating on Netflix')
plt.subplot(122)
df['rating'].value_counts()[:10].plot(kind='pie',autopct='%.2f%%') # plotting the top 10 ratings
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.title('Rating')
plt.show()
```

* TOP COUNTRIES WHOSE MOVIES ARE ADDED IN THE NETFLIX

```python
print('Top 10 countries to release movies and shows:\n',df['country'].value_counts()[0:10])

#plotting the top 10 countries
plt.subplots(figsize=(15,5))
plt.subplot(121)
df['country'].value_counts()[:10].plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.title('Country')
plt.subplot(122)
df['country'].value_counts()[:10].plot(kind='pie',autopct='%.2f%%') # plotting the top 10 countries in production wise
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.title('country')
plt.show()
```

* TOP 10 DIRECTORS TO PRODUCDE MOVIES

```python
print('Top 10 directors to release movies on Netflix:\n',df['director'].value_counts()[1:10])

# Top 10 directors whose movies are added // Excludd the movies and TV Shows whose directors name is not given
plt.subplots(figsize=(15,5))
plt.subplot(121)
df['director'].value_counts()[1:10].plot(kind='bar')
plt.xlabel('Directors')
plt.ylabel('Frequency')
plt.title('Director')
plt.subplot(122)
df['director'].value_counts()[1:10].plot(kind='pie',autopct='%.2f%%') # plotting the top 10 countries in production wise
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.title('Director')
plt.show()
```

* PLOTTING THE TOP 10 GENRE WITH THEIR FREQUENCY

* converting the date_added column to the datetime datatype

```python
df['date_added']=pd.to_datetime(df['date_added'])
df['added_year']=df['date_added'].dt.year # Extracting the year
df['added_month']=df['date_added'].dt.month # Extracting the month
df['added_month_name']=df['date_added'].dt.strftime('%b')
df['day_name']=df['date_added'].dt.day_name()  # Extracting the day of the week
df['day_is_weekened']=np.where(df['day_name'].isin(['Sunday','Saturday']),1,0)  # checking whether it is weekened or not 
df.sample(5)
```

* checking whether the relase date is a weekened or not?

```python
df['day_is_weekened'].value_counts().plot(kind='bar') # 1 means weekened whether 0 means not weekened 
```

```python
counts = df.groupby(['day_name', 'type']).size().unstack(fill_value=0)

# Reorder the days 
order_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
counts = counts.reindex(order_days)


# Plot the bar chart
counts.plot(kind='bar', figsize=(10, 5))
plt.title('Types Vs days in a week')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.tight_layout()
plt.show()
```

* PLOTTING THE MOVIES AND TVSHOWS RELEASE WITH THE MONTHS

```python
counts = df.groupby(['added_month_name', 'type']).size().unstack(fill_value=0)

# Reorder the days if needed
order_month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
counts = counts.reindex(order_month)

# Plot the bar chart
counts.plot(kind='bar',figsize=(10,5))
plt.title('Types VS Month')
plt.xlabel('Months')
plt.ylabel('Count')
plt.legend(title='Type')
plt.show()
```
